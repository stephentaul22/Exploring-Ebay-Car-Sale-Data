There are a lot of outliers in the price column. I will begin by removing the cars priced at 0, and then discover which cars to eliminate afterward.

```python
autos_nonfree = autos[autos['price'] != 0]
autos_nonfree.info()
```

I will use the Tukey Fences method to find where to draw the cutoff points for outliers.

```python
Q1 = autos_nonfree['price'].quantile(0.25)
Q3 = autos_nonfree['price'].quantile(0.75)
IQR = Q3 - Q1

autos_filtered = autos_nonfree.query('(@Q1 - 1.5 * @IQR) <= price <= (@Q3 + 1.5 * @IQR)')
autos_filtered.info()
```

```python
autos_filtered['brand'].value_counts()
```

```python
price_share_per_brand = {}

for brand in autos_filtered['brand'].unique():
    autos_brand = autos_filtered[autos_filtered['brand'] == brand]
    price_share_per_brand[brand] = round(autos_brand['price'].sum()/autos_filtered['price'].sum() * 100, 2)
    
price_share_per_brand
```

```python
price_share_brand_sorted = {}

for i in sorted(price_share_per_brand.values(), reverse=True):
    for j in price_share_per_brand.keys():
        if price_share_per_brand[j] == i:
            price_share_brand_sorted[j] = price_share_per_brand[j]
            break

price_share_brand_sorted
```

## Analysis
6 Brands have a a market share of 5% or greater. I will now focus only on these brands.

```python
brands = []
avg_price_brand = {}

for key in price_share_brand_sorted.keys():
    brands.append(key)

for brand in brands[:6]:
    avg_price_brand[brand] = round(autos_filtered[autos_filtered['brand'] == brand]['price'].mean(), 2)

avg_price_brand_sorted = {}
sorted_keys = sorted(avg_price_brand, key=avg_price_brand.get, reverse=True)

for key in sorted_keys:
    avg_price_brand_sorted[key] = avg_price_brand[key]

avg_price_brand_sorted
```

## Analysis
While Volkswagon had the highest market share, its average price per car was 4th highest. Audi had the highest price per car of the 6 brands, followed closely by BMW and Mercedes Benz. Ford and Opel had much small average prices compared to the other 4 brands in the top 6 market share.

```python
avg_mileage_brand = {}

for brand in brands[:6]:
    avg_mileage_brand[brand] = round(autos_filtered[autos_filtered['brand'] == brand]['kilometer'].mean(), 2)

avg_mileage_brand_sorted = {}
sorted_keys = sorted(avg_mileage_brand, key=avg_mileage_brand.get, reverse=True)

for key in sorted_keys:
    avg_mileage_brand_sorted[key] = avg_mileage_brand[key]

avg_mileage_brand_sorted
```

```python
avg_mileage = pd.Series(avg_mileage_brand_sorted)
avg_price = pd.Series(avg_price_brand_sorted)
top_brands_averages = pd.DataFrame(avg_price)
top_brands_averages['average price'] = top_brands_averages
top_brands_averages['average mileage'] = avg_mileage
top_brands_averages = top_brands_averages[['average price', 'average mileage']]
top_brands_averages
```

## Conclusion
After cleaning the data significantly, Audi comes out the clear winner on the Ebay market in Germany. 5 of the top 6 brands on the Ebay market in Germany are german. The only brand breaking this pattern is Ford, from the United States. There is a generally direct relationship between average price of a car by brand and average mileage of a car by brand. A couple brands do avoid the trend though.
